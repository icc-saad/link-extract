```python
import pandas as pd 
```

```python
df = pd.read_csv('./undp popp docs/POPP_Metadata.csv')
```

```python
df['Document Language'].value_counts().to_frame()
```

```python
def get_filetype(filename):
    return filename.split('.')[-1].lower()

baseURL = 'https://popp.undp.org/UNDP_POPP_DOCUMENT_LIBRARY/Public/'
def get_url(filename):
    return baseURL + filename
```

```python
df['FileType'] = df['Name'].apply(get_filetype)
```

```python
df['URL'] = df['Name'].apply(get_url)
```

```python
df['FileType'].value_counts().to_frame()
```

```python
df.to_csv('./undp popp docs/POPP_Metadata + url and type.csv', index=False)
```
